```python
#Recreate Graph Week 4
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.ticker as ticker

from matplotlib.dates import DateFormatter
from matplotlib.pyplot import figure

filename = '/content/athleteEventsNoPersonal (4).csv'

olympics_df = pd.read_csv(filename)
```

```python
# How many Gold medals were awarded to athletes who were over 40 years old? 
# creating goldmedal dataframes
goldMedals = olympics_df[(olympics_df == 'Gold')]
print(goldMedals.head())

goldMedals = olympics_df[(olympics_df.Medal == 'Gold')]
print('The no of athletes is',
      goldMedals['ID'][goldMedals['Age'] > 40].count(), '\n')
print(goldMedals[goldMedals['Age'] > 40])

masterDisciplines = goldMedals['Sport'][goldMedals['Age'] > 50]
plt.figure(figsize=(20, 10))
plt.tight_layout()
sns.countplot(masterDisciplines)
plt.title('Gold Medals for Athletes Over 40')
plt.show()
```
